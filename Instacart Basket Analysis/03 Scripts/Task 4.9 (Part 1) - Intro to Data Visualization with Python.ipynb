# 03. Data quality and consistency checks

```python
# Check for missing values
df_cust.isnull().sum()
```

```python
df_cust[df_cust['first_name'].isnull() == True]
```

#### Although there are 11,259 rows with missing values for the first name, we can still identify them by their user id

```python
# Check for duplicates
df_cust[df_cust.duplicated()]
```

#### No duplicates were found

```python
# Check for mixed types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# Change the dataype of first_name to type string (object)
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
# Check for mixed types again
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

#### No more mixed type data
